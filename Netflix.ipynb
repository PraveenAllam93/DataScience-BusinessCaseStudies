### handling rating missing data

#### Imputing rating missing data

```python
#Finding the unique rating values present
netflix_data['rating'].unique()
```

```python
# finding the data which are missed with rating values 
rating_missing_data = netflix_data[pd.isna(netflix_data['rating'])]
rating_missing_data.head()
```

```python
# droping the missing rating data from the main dataframe
netflix_data = netflix_data.drop(index = rating_missing_data.index)
netflix_data
```

```python
# information about the rating missed data
rating_missing_data.info()
```

```python
# finding the country's present in the data, to impute the rating.
missing_rating_country = rating_missing_data[~(rating_missing_data['country'] == 'nan')]['country'].unique().tolist()
missing_rating_country
```

```python
# finding the genre present in the data, to impute the rating
missing_rating_genre = rating_missing_data['listed_in'].unique().tolist()
missing_rating_genre
```

```python
#extracting the required countries
country_mask = np.isin(netflix_data['country'], missing_rating_country)
data_in_given_countries_genre = netflix_data[(country_mask)]

# extracting the required genre's
genre_mask = np.isin(data_in_given_countries_genre['listed_in'], missing_rating_genre)
data_in_given_countries_genre = data_in_given_countries_genre[(genre_mask)]

#getting the required masked data
rating_null = pd.isna(data_in_given_countries_genre['rating'])
data_in_given_countries_genre = data_in_given_countries_genre[~(rating_null)]
data_in_given_countries_genre = data_in_given_countries_genre[['title', 'listed_in', 'country','rating']]
data_in_given_countries_genre
```

```python
# grouping by genre & country and finding the max rating in the each pool
max_rating = data_in_given_countries_genre.groupby(['listed_in', 'country'])['rating'].agg('max')
max_rating.reset_index()
```

```python
# finding max rating for respective genre
max_rating_with_no_country = netflix_data.groupby(['listed_in'])['rating'].agg('max')
max_rating_with_no_country.reset_index()
```

```python
# dropping the rating column
rating_missing_data.drop('rating',axis = 1, inplace = True)
rating_missing_data.head()
```

```python
# merging the data
rating_missing_data_no_country = rating_missing_data[rating_missing_data['country'] == 'nan']
rating_missing_data_no_country = rating_missing_data_no_country.merge(max_rating_with_no_country, on = 'listed_in', how = 'inner')
rating_missing_data_no_country
```

```python
# merging the data
rating_missing_data = rating_missing_data.merge(max_rating, on = ['listed_in', 'country'], how = 'inner')
rating_missing_data
```

```python
# concating the rating data's
rating_data = pd.concat([rating_missing_data,rating_missing_data_no_country])
rating_data
```

```python
# concating the main df and imputed rating data
netflix_data = pd.concat([netflix_data, rating_data])
```

```python
netflix_data = netflix_data.reset_index()
netflix_data = netflix_data.drop('index', axis = 1)
```
